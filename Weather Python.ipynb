# Checking the Correlation

```python
 # Visualizing the Correlation Matrix

correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Weather features")

plt.ylabel("Weather features")

plt.show()
```

```python
# Ordering the dataset by Rel Hum_% column

pd.set_option('display.max_rows', 20)

df.sort_values(by=['Rel Hum_%'], inplace=False, ascending=True)
```

```python
# Comparing the Temp_C and the Press_kPa using a scatter plot

plt.scatter(x=df['Temp_C'], y=df['Press_kPa'])

plt.title('Temperature vs Pressure')

plt.xlabel('Temperature')

plt.ylabel('Pressure')

plt.show
```
